# Location Profiling with H3 Python API

#ref
[1] https://betterprogramming.pub/playing-with-ubers-hexagonal-hierarchical-spatial-index-h3-ed8d5cd7739d

```python
from h3 import h3
import folium

lat=13.675813230683476
lng= 100.55001984172273
m = folium.Map(location=[lat, lng], zoom_start=11, tiles='cartodbpositron')

resolution=8
hexagons1=[]
hexagons1.append(h3.geo_to_h3(lat, lng, resolution))
kRing = h3.k_ring(hexagons1[0], 1)
hexagons1=hexagons1+list(kRing)

polylines = []
for hex in hexagons1:
    polygons = h3.h3_set_to_multi_polygon([hex], geo_json=False)
    # flatten polygons into loops.
    outlines = [loop for polygon in polygons for loop in polygon]
    polyline = [outline + [outline[0]] for outline in outlines][0]
    polylines.append(polyline)
for polyline in polylines:
    my_PolyLine=folium.PolyLine(locations=polyline,weight=4,color='red')
    m.add_child(my_PolyLine)
display(m)

```
